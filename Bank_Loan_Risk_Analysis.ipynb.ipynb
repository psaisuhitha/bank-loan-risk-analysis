```python
plt.figure(figsize=(6,4))
sns.histplot(df['Income'], kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include=['number']).corr(), 
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
//The analysis shows that DTIRatio has a strong positive correlation with Default, 
indicating that customers with higher debt-to-income ratios are more likely to default.
```

```python
//CreditScore shows a negative correlation with Default, indicating that customers with lower credit scores are more likely to default,
  while higher credit scores reduce credit risk
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Default', y='CreditScore', data=df)
plt.title("Credit Score vs Default")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Income', y='LoanAmount', hue='Default', data=df)
plt.title("Income vs Loan Amount")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Default', y='DTIRatio', data=df)
plt.title("DTI Ratio vs Default")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Default', y='Income', data=df)
plt.title("Income vs Default")
plt.show()
```

```python
//The median credit score for defaulters is significantly lower compared to non-defaulters,
indicating that lower credit scores strongly increase the probability of loan default.
```

```python
df['Advanced_Risk'] = np.where(
    (df['CreditScore'] < 600) |
    (df['DTIRatio'] > 0.4) |
    (df['Income'] < 30000),
    "High Risk",
    "Low Risk"
)

df['Advanced_Risk'].value_counts()
```

```python
pd.crosstab(df['Advanced_Risk'], df['Default'], normalize='index')
```

```python
sns.countplot(x='Advanced_Risk', hue='Default', data=df)
plt.title("Risk Category vs Default")
plt.show()
```

```python
# Key Insights:
# 1. CreditScore negatively correlates with Default.
# 2. DTIRatio positively correlates with Default.
# 3. Lower income customers show higher default rates.
# 4. Advanced_Risk segmentation effectively separates high-risk borrowers.
```
